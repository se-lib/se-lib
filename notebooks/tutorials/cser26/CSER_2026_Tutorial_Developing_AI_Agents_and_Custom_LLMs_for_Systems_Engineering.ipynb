## Introduction

### GenAI Terminology

- **Large Language Models (LLMs)**: Neural networks to
understand and generate text and other forms of data.
- **Generative Pre-trained Transformer (GPT)**: A specific type of LLM architecture that uses the transformer neural network, is pre-trained on large text datasets, and generates text by predicting the next token in sequence.
- **Agent** - An AI-enabled software entity that performs a defined role using context, instructions, and available interfaces such as prompts, tools, APIs, or workflows. An agent need not be fully autonomous.
- **Training**: Teaching a model from scratch how to perform
a task using large amounts of data.
- **Retraining**: Retraining a base model with a new dataset.
- **Finetuning**: Teaching a base model specialized knowledge
by adjusting internal parameters such as weights and
biases.
- **Retrieval Augmented Generation (RAG)**: Utilizing
external data to augment an LLM’s knowledge without
changing model parameters.
- **Pipeline**: A series of connected stages that transform data, interact with an LLM, and produce a desired output
- **Tailoring**: Adapting or customizing a general-purpose LLM to perform specific tasks or generate specialized outputs for a particular domain or application.
- **Reinforcement Learning from Human Feedback (RLHF)**:
An iterative technique used to train AI models by
incorporating human feedback to improve their
performance and alignment with human preferences.

# AI Integration Using APIs is Bidirectional

- There are two complementary API integration patterns in systems engineering
- In AI-to-tool integration, the AI assistant calls external tools or services
- In tool-to-AI integration, an external process or enterprise application calls the AI model


# AI Calling External Tools

- The AI interprets the user request and decides when to invoke another tool
- External tools provide authoritative data, formal computation, simulation, optimization, or rules-based analysis


# External Processes Calling AI

- An enterprise workflow, engineering application, or automation layer calls the AI model through an API
- The host process remains in control of logic, sequencing, data, security, and validation
- The AI contributes summarization, extraction, explanation, classification, drafting, or reasoning support

#### Constrain LLM Output for Control System Component Classification

````python
# Constrain LLM Output for Control System Component Classification
from google import genai
from google.colab import userdata
import enum

# Define a comprehensive enumeration for Industrial Control System (ICS) Component Types
class IcsComponentType(enum.Enum):
    VALVE = "Valve"
    SENSOR_TEMPERATURE = "Temperature Sensor"
    SENSOR_PRESSURE = "Pressure Sensor"
    ACTUATOR_MOTOR = "Motor Actuator"
    ACTUATOR_SOLENOID = "Solenoid Actuator"
    CONTROLLER_PLC = "PLC Controller"
    CONTROLLER_DCS = "DCS Controller"
    COMM_GATEWAY = "Communication Gateway"
    POWER_SUPPLY = "Power Supply Unit"
    HMI_PANEL = "Human-Machine Interface Panel"
    NETWORK_SWITCH = "Network Switch"
    CIRCUIT_BREAKER = "Circuit Breaker"
    PUMP = "Pump"
    FLOW_METER = "Flow Meter"
    TRANSDUCER = "Transducer"

client = genai.Client(api_key=userdata.get('gem'))

# Example query: A natural language description of a faulty component
component_description = "electricity come from it"

response = client.models.generate_content(
    model='gemini-2.0-flash', # Or a more capable model if needed for complex reasoning
    contents=f'Classify the following industrial component based on its function: "{component_description}"',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': IcsComponentType, # Constrain output to our defined enum
    },
)

# Print the classified component type
print(f"Original description: \"{component_description}\"")
print(f"Classified Component Type: {response.text}")

# Example of how a downstream system might use this:
if response.text == IcsComponentType.SENSOR_TEMPERATURE.value:
    print("Action: Initiate temperature sensor diagnostics protocol and check spare parts inventory for 'THERM-001'.")
elif response.text == IcsComponentType.VALVE.value:
    print("Action: Check valve actuator status and prepare for mechanical inspection.")
else:
    print("Action: Refer to general component troubleshooting guide.")
````

#### ✅ Exercise - LLM Classification

````python
# Tailored LLM for Systems Engineering Project Risk Classification
from google import genai
from google.colab import userdata
import enum

# Define a comprehensive enumeration for common Systems Engineering Project Risk Categories
class SeProjectRiskCategory(enum.Enum):
    TECHNICAL = "Technical Risk"
    SCHEDULE = "Schedule Risk"
    COST_FINANCIAL = "Cost/Financial Risk"
    REQUIREMENTS_MANAGEMENT = "Requirements Management Risk"
    RESOURCE_PERSONNEL = "Resource/Personnel Risk"
    STAKEHOLDER_COMMUNICATION = "Stakeholder/Communication Risk"
    EXTERNAL_REGULATORY = "External/Regulatory Risk"
    SUPPLY_CHAIN_PROCUREMENT = "Supply Chain/Procurement Risk"
    PERFORMANCE_QUALITY = "Performance/Quality Risk"
    OPERATIONAL_TRANSITION = "Operational/Transition Risk"
    GOVERNANCE_MANAGEMENT = "Governance/Management Risk"

client = genai.Client(api_key=userdata.get('gem'))

# Example query: A natural language description of a project risk
# Scenario 1: Ambiguous technical risk
risk_description_1 = "We're struggling with integrating the new radar module with the existing navigation system. Data formats don't match, and the APIs are undocumented."

# Scenario 2: Schedule risk due to external dependency
risk_description_2 = "The delivery of the specialized test equipment from Vendor X is delayed by 3 months, which impacts our integration test phase start date."

# Scenario 3: Resource/Personnel risk
risk_description_3 = "Our lead systems architect announced their retirement, and we don't have a clear successor with the same institutional knowledge."

# Scenario 4: Cost risk
risk_description_4 = "The projected cost of the custom-fabricated housing for the satellite exceeds initial estimates by 25% due to raw material price increases."

# --- Process Scenario 1 ---
print(f"--- Analyzing Risk 1 ---")
print(f"Original description: \"{risk_description_1}\"")
response_1 = client.models.generate_content(
    model='gemini-2.0-flash', # Or a more capable model if needed for complex reasoning
    contents=f'Classify the following systems engineering project risk: "{risk_description_1}"',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': SeProjectRiskCategory, # Constrain output to our defined enum
    },
)
print(f"Classified Risk Category: {response_1.text}")
if response_1.text == SeProjectRiskCategory.TECHNICAL.value:
    print("Action: Assign to Engineering Lead, review interface control documents, schedule technical deep dive.")
print("\n")

# --- Process Scenario 2 ---
print(f"--- Analyzing Risk 2 ---")
print(f"Original description: \"{risk_description_2}\"")
response_2 = client.models.generate_content(
    model='gemini-2.0-flash',
    contents=f'Classify the following systems engineering project risk: "{risk_description_2}"',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': SeProjectRiskCategory,
    },
)
print(f"Classified Risk Category: {response_2.text}")
if response_2.text == SeProjectRiskCategory.SCHEDULE.value:
    print("Action: Update critical path, inform stakeholders, explore alternative suppliers.")
print("\n")

# --- Process Scenario 3 ---
print(f"--- Analyzing Risk 3 ---")
print(f"Original description: \"{risk_description_3}\"")
response_3 = client.models.generate_content(
    model='gemini-2.0-flash',
    contents=f'Classify the following systems engineering project risk: "{risk_description_3}"',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': SeProjectRiskCategory,
    },
)
print(f"Classified Risk Category: {response_3.text}")
if response_3.text == SeProjectRiskCategory.RESOURCE_PERSONNEL.value:
    print("Action: Initiate knowledge transfer plan, identify potential internal/external candidates, develop succession plan.")
print("\n")

# --- Process Scenario 4 ---
print(f"--- Analyzing Risk 4 ---")
print(f"Original description: \"{risk_description_4}\"")
response_4 = client.models.generate_content(
    model='gemini-2.0-flash',
    contents=f'Classify the following systems engineering project risk: "{risk_description_4}"',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': SeProjectRiskCategory,
    },
)
print(f"Classified Risk Category: {response_4.text}")
if response_4.text == SeProjectRiskCategory.COST_FINANCIAL.value:
    print("Action: Review budget, identify cost-reduction opportunities, request budget re-allocation if necessary.")
print("\n")
````

#### Structured Prompt for System Network Configuration Diagram

````python
# Structured Prompt for System Network Configuration Diagram
from google import genai
# Assuming `google.colab.userdata` is available in your environment,
# otherwise, replace with a secure method to get API key (e.g., os.getenv)
# from google.colab import userdata

import os
import graphviz # pip install graphviz

# --- Client Setup (using environment variable for better practice outside Colab) ---
# For Colab, you might still use userdata.get('gem')
# For local development, it's recommended to set GEMINI_API_KEY environment variable
api_key = os.getenv('GEMINI_API_KEY')
if not api_key:
    # Fallback to userdata if running in Colab and env var isn't set
    # Or, prompt user with getpass if not in Colab and env var is missing
    try:
        from google.colab import userdata
        api_key = userdata.get('gem')
        if not api_key:
            raise ValueError("Gemini API key not found in Colab userdata.")
    except (ImportError, ValueError):
        import getpass
        print("Gemini API key not found in environment variable 'GEMINI_API_KEY'.")
        api_key = getpass.getpass("Please enter your Gemini API key: ")

client = genai.Client(api_key=api_key)

# --- Systems Engineering Network Diagram Prompt ---
se_network_prompt = """
TASK:
Generate a Graphviz DOT language script for a system network configuration diagram.
The diagram should show the specified SYSTEM COMPONENTS as nodes and their direct COMMUNICATION LINKS as edges.
Use distinct shapes or colors for different types of components if applicable.

SYSTEM COMPONENTS:
- Main Server (Type: Server, Role: Central Processing)
- Database Server (Type: Server, Role: Data Storage)
- Web Server (Type: Server, Role: User Interface)
- Load Balancer (Type: Network Device, Role: Traffic Distribution)
- Firewall (Type: Security Device, Role: Network Security)
- Client Workstation (Type: End Device, Role: User Access)
- External Service Gateway (Type: Network Device, Role: External Integration)

COMMUNICATION LINKS:
- Client Workstation -> Load Balancer (HTTPS)
- Load Balancer -> Web Server (HTTP)
- Web Server -> Database Server (JDBC/ODBC)
- Web Server -> Main Server (Internal API)
- Main Server -> Database Server (Internal API)
- Main Server -> External Service Gateway (REST API)
- External Service Gateway -> Main Server (REST API)
- Firewall sits between Load Balancer and Web Server.
- All internal servers (Main, Database, Web) are on a secure internal network segment.

Graphviz Specifics:
- Use a `digraph` for directed connections.
- Label edges with the communication protocol/type.
- Use distinct node shapes for Servers (e.g., box or cylinder), Network Devices (e.g., triangle or diamond), Security Devices (e.g., octagon), and End Devices (e.g., ellipse).
- Cluster internal servers within a subgraph to represent the "secure internal network segment".
"""

print("Generating Graphviz DOT for network diagram...")
response = client.models.generate_content(
    model='gemini-2.0-flash', # gemini-1.5-pro might be even better for complex structures
    contents=se_network_prompt
)

dot_code = response.text
print("\n--- Generated Graphviz DOT Code ---")
print(dot_code)
````

# Integrating LLMs with External Tools Through APIs

# Model Evaluation with Benchmarking

## Background on Benchmarking Architecture

A **Model** (identified by a name and version) can be evaluated on zero or more **Benchmarks**.

Each **Benchmark** (given a name and a domain such as “general NLP,” “reasoning,” or “safety”) contains one or more **Tasks**.

A **Task** (e.g., QA, summarization, translation, coding) specifies what the model must do and its modality (text-to-text, code-generation, etc.).

Finally, every **Task** is measured by one or more **Metrics** that spell out exactly how performance is scored (e.g., Accuracy, ROUGE-L, BLEU), including a formula or calculation rule.

The cardinalities on the connectors (1, 0.., 1..) make explicit that a model may face several benchmarks, each benchmark must include at least one task, and each task must be paired with at least one metric.

image.png

#### **Other relevant terms and definitions:**

Tensors are the primary way language models (LLMs) represent and manipulate data — from input tokens to internal states and final predictions. Every stage of an LLM pipeline, including tokenization, embedding, attention, and decoding, involves operations on tensors of varying dimensions.

A tensor is a multi-dimensional array of numerical values. It generalizes:
* Scalars (0D) → A single number (e.g., loss = 0.32)
* Vectors (1D) → A list of numbers (e.g., token logits)
* Matrices (2D) → A table of values (e.g., embeddings)
* Higher-order tensors (3D+) → Used extensively in LLMs for batching, attention, and more.

## Background on lm-evaluation-harness

#### **What is lm-evaluation-harness?**
The **lm-evaluation-harness** is an open-source test-rig from EleutherAI that wraps dozens of popular NLP benchmarks behind a single, uniform interface.

It handles the unglamorous plumbing—prompt templating, batching, sampling, metric computation, and result aggregation—so you can point any Hugging Face (or API-served) model at a task and get a tidy JSON/CSV score report.

Because every task is expressed in the same declarative YAML/JSON format, swapping models or adding evaluation flags (e.g., few-shot vs. zero-shot, greedy vs. sampled decoding) is as simple as changing a CLI argument, which makes large comparison grids and automated regression tests straightforward and reproducible.

#### **lm-evaluation-harness Structure**

First, you provide two things: a set of benchmark tasks—like MMLU or GSM8K—and your language model, whether it's GPT, Claude, or LLaMA.

These go into the Evaluation Engine, the core of the system.

The engine uses built-in tools to streamline the process:

Smart caching avoids re-running identical requests.

A metrics calculator computes scores like accuracy, F1, or BLEU.

And a rich task library gives you 50+ prebuilt evaluations out of the box (e.g. other benchmarks).

Once done, the harness returns detailed results complete with scores, comparisons, and breakdowns—so you can deeply analyze how your model performs.

image.png

#### **lm-evaluation-harness Behavior**
A simple 3-step process to benchmark a model on the MMLU dataset.

**Step 1: Initialize Tasks and Model.**
Run a single command to load 14,000 multiple-choice questions and connect to your model.

**Step 2: Perform Batch Evaluation.**
The system feeds each question to your model, checks the answer, and stores the result. This runs automatically for all questions.

**Step 3: Analyze and Score Results.**
The harness calculates the model’s accuracy and gives you a detailed report by subject.

For example, you might get an MMLU score of 67.2%.

In just three steps, you’ll know exactly how your model performs across a wide range of academic and professional topics.

image.png

#### **Why lm-evaluation-harness over other tools?**

**lm-evaluation-harness** stands out because it lets us do rigorous, research-grade benchmarking entirely on our own hardware, with our own models, and for any task or metric we care to define.

Unlike OpenAI Evals, it is completely open-source, works offline, and is not tied to a single vendor’s API; unlike HF Evaluate, it handles the whole pipeline—prompting, inference, and scoring—rather than just the metric math; and unlike Chatbot Arena, it delivers deterministic, objective numbers instead of subjective crowd votes.

In practice that means we get broad structured-task coverage (MCQs, fill-in-the-blank, logic, math, etc.), fully automatic and reproducible scoring, and plug-and-play hooks for custom JSON/CSV datasets or bespoke metrics—all in a modular CLI that already speaks Hugging Face, GGUF, and plain HTTP APIs.

In short, lm-evaluation-harness is the only option in the comparison that combines scientific rigor, vendor independence, and turnkey extensibility, making it the clear choice for systematic benchmarking research on our custom benchmark.

#### **Language Model evaluation taxonomy task Versus lm-evaluation-harness task**

In the **taxonomy**, a **task** is *the thing being tested*—a coherent slice of a benchmark such as “multiple-choice QA on MMLU” or “document summarization on XSum.” It describes the content and modality of the questions.

In **lm-evaluation-harness**, a **task** is *the recipe for running that test*: a YAML file that tells the harness where the data live, how to prompt the model, and which metrics to score.

So the taxonomy’s task is an **evaluation target**, while the harness’s task is an executable **job definition** for hitting that target.

#### **Creating an lm-evaluation-harness task**

We supply is a **small, plain-text YAML file** that lists where the dataset lives, how each row should be turned into a prompt and an answer key, and which metric to log. Dropping that YAML into the harness’ tasks/ folder instantly turns our dataset into a “task” that can be evaluated with a single command (lm_eval …).

Because all the logic sits in the YAML rather than in code, *non-programmers can set up or modify tests simply by editing text, and the resulting scores are fully reproducible across machines, models, and teams*. This lets us compare our custom benchmark against public suites (MMLU, XSum, etc.) without writing bespoke scripts, keeping the evaluation process transparent and easy to maintain.

### Evaluation Metrics

Metrics help assess how well models perform based on different criteria. There are different categories of criterion, including:
* Exact-match / Classification: Measures correctness (e.g., Accuracy).
* Similarity: Measures overlap or closeness to reference (e.g., BLEU, ROUGE).
* Scoring: Evaluates fluency, coherence, factuality (e.g., Perplexity).
* Behavioral: Measures robustness, fairness, safety (e.g., toxicity, bias).

| **Metric Type**      | **Use Case**                      | **Limitations**                                 |
| -------------------- | --------------------------------- | ----------------------------------------------- |
| **Accuracy**         | Classification, Multiple Choice   | Ignores partial correctness                     |
| **BLEU / ROUGE**     | Text similarity (e.g., summaries) | Penalizes paraphrasing; limited fluency capture |
| **BERTScore**        | Semantic similarity               | May favor fluent but factually wrong outputs    |
| **Human Evaluation** | Ground-truth utility assessments  | Expensive, slow, inconsistent                   |

### Types of Output Files

`samples.jsonl` tells you why the model got something right or wrong; `results.json` tells you how well it did overall. Parse the former into a DataFrame for deep dives and trust the latter for quick reporting.

#### **Results Files**

`results.json` files gives you the headline numbers for each task—typically overall accuracy, normalized accuracy, and any other metric you requested. Treat this file as the scoreboard: load it with json.load() and print or plot the values to compare runs. Because the file is already aggregated, you can drop it straight into a dashboard or a Markdown report; if two models produce the same set of metrics, a side‑by‑side bar chart or simple table instantly shows which system wins on each score.

#### **Samples Files**

`samples.json`, by contrast, is the diagnostic goldmine. Each line is a full JSON record for one example, containing the prompt, all answer choices, the gold label, the model’s prediction, and (if possible) per‑choice log‑probabilities.
